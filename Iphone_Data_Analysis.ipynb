```python
figure = px.bar(df,
               title="Daily App Usage",
               x="Date",
               y="Usage",
               color="App")
figure.show()
```

```python
# Create a pie chart
fig = px.pie(df, values='Usage', names='App',title =" Pie Chart Showing App Usage By Age")

# Display the pie chart
fig.show()
```

```python
figure = px.scatter(df,
                    x="Notifications",
                    y="Usage",
                    size="Notifications",
                    trendline="ols",
                    title = "Relationship Between Number of Notifications and Usage")
figure.show()
```

```python
apps = df['App'].value_counts()
plt.figure(figsize=(15,8))
plt.bar(apps.index, apps, color= ['#A0C4FF','#F7A76C'])
plt.xticks(rotation = 60)
plt.show()
```
